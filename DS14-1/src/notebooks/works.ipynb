```python
import pandas as pd
import ast
```

```python
def parse(parsed_column):
    L = []
    for i in ast.literal_eval(parsed_column):
        L.append(i['name'])
    return L
```

```python
distance = pd.read_csv('../assets/distance.csv', index_col='id')
distance.index = distance.index.astype(int)
distance.columns = distance.columns.astype(int)

movies_all = pd.read_csv('../assets/movies.csv', index_col='id')
movies_all.index = movies_all.index.astype(int)
movies_all['genres'] = movies_all['genres'].apply(parse)
movies_all['years'] = pd.DatetimeIndex(
    movies_all.release_date).year.fillna(0).astype(int)
movies = movies_all
```

```python
def recommendation(title: str, top_k: int = 5):
    # filtered = self.movies.genres
    indices = pd.Series(
        movies_all.index, index=movies_all['title']).drop_duplicates()
    print(indices)
    idx = indices[title]
    sim_scores = list(enumerate(distance[idx]))
    print(sim_scores)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    print(sim_scores)
    sim_scores = sim_scores[1:top_k+1]
    print(sim_scores)
    movie_ind = [i[0] for i in sim_scores]
    print([indices[i] for i in movie_ind])
    print(idx, indices[0])
    return list(movies_all['title'].iloc[movie_ind])
```

```python
recommendation('Avatar', 5)
```
