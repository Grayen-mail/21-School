## Задание 6
Объедини 2 предобработанных датасета. Результат запиши в переменную `X_ready`. Сохрани датасет в папку `datasets/data_prepared.csv`
Подели получившийся датасет на `train` и `test`. Процент тестовой выборки: `20%`. Использовать параметр `random_state=21`.
Выведи размерность обучающей выборки X_train.


```python
# Соединим по строкам, т.е. вширь - wide (axis=1)
X_ready = pd.concat([X_num_scal1, X_cat_ohe], axis=1)
# Количество столбцов 370 = 323 + 47
X_ready.shape, X_ready
```

```python
# Запишем получившийся датасет в файл
X_ready.to_csv("../datasets/data_prepared.csv", index=False)
```

```python
# Разделение датасета на две выборки - обучающую и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(  # в соотношении 80%/20%, и задаем конкретное начальное значение
    X_ready, Y, test_size=0.2, random_state=21)       # псевдогенератора случайных чисел, для повторения выборки
X_train.shape, X_train, X_test.shape
```

## Задание 7 
Обучи логистическую регрессию с параметрами: `solver='liblinear'`, `fit_intercept=False`, `penalty='l1'`, `random_state=21`. 
Посчитай `accuracy` модели на тестовой выборке, сравнив предсказанные значения с реальными.
Выведи значение `accuracy` модели на тестовой выборке.

```python
# Функция классификации для прогнозирования вероятности по двоичным данным
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear', fit_intercept=False,  # т.е. обучение модели
                         penalty='l1', random_state=21).fit(X_train, Y_train)
clf.coef_
```

```python
# Прогнозирование результатов на тестовой выборке, на основе обученной модели
pred = clf.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, pred).round(
    6)  # расчет точности предсказания модели
accuracy
```

## Задание 9
Построй график, на котором видны топ-10 самых важных факторов по мнению модели в абсолютном значении.

```python
# Приведем к абсолютному значению и извлечёем из внешнего списка
abs_coef = clf.coef_.__abs__()[0]
# Отсортируем по значению и отберем индексы последних 10 элементов
abs_index_top = abs_coef.argsort()[-10:]
# По индексам выберем значения коэффициентов ...
abs_top = abs_coef[abs_index_top]
coef_name = X_ready.columns[abs_index_top]  # ... и названия столбцов
coef_name
```

```python
import seaborn as sns
# И посмотрим, что получилось
ax1 = sns.barplot(y=coef_name, x=abs_top, palette='colorblind', width=0.8)
ax1.set_xlabel('Абсолютное значение коэффициента')
ax1.set_ylabel('Факторы')
ax1.set_title('ТОП-10 важных факторов')
```
